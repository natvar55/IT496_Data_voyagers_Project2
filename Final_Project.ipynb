Our Dataset is IP Network Traffic Flows Labeled with 75 Apps.We have 87 features and 3577296 instances.

Flow.ID: A unique identifier for each network flow, which represents a sequence of packets between a source and destination.

Source.IP: The IP address of the source (sender) in the network flow.

Source.Port: The port number on the source side used for the communication.

Destination.IP: The IP address of the destination (receiver) in the network flow.

Destination.Port: The port number on the destination side used for the communication.

Protocol: The network protocol used for communication (e.g., TCP, UDP, ICMP).

Timestamp: The timestamp when the flow was recorded.

Flow.Duration: The duration of the flow, which is the time it took for the communication to complete.

Total.Fwd.Packets: The total number of packets sent from the source to the destination.

Total.Backward.Packets: The total number of packets sent from the destination back to the source.

Total.Length.of.Fwd.Packets: The total length (in bytes) of all forward packets.

Total.Length.of.Bwd.Packets: The total length (in bytes) of all backward packets.

Fwd.Packet.Length.Max: The maximum length of a forward packet.

Fwd.Packet.Length.Min: The minimum length of a forward packet.

Fwd.Packet.Length.Mean: The mean (average) length of forward packets.

Fwd.Packet.Length.Std: The standard deviation of the lengths of forward packets.

Bwd.Packet.Length.Max: The maximum length of a backward packet.

Bwd.Packet.Length.Min: The minimum length of a backward packet.

Bwd.Packet.Length.Mean: The mean (average) length of backward packets.

Bwd.Packet.Length.Std: The standard deviation of the lengths of backward packets.

Flow.Bytes.s: The rate of data transfer in bytes per second for the entire flow.

Flow.Packets.s: The rate of packet transfer per second for the entire flow.

Flow.IAT.Mean: The mean inter-arrival time (time between consecutive packets) for the flow.

Flow.IAT.Std: The standard deviation of the inter-arrival times for the flow.

Flow.IAT.Max: The maximum inter-arrival time for the flow.

Flow.IAT.Min: The minimum inter-arrival time for the flow.

Fwd.IAT.Total: The total inter-arrival time for forward packets.

Fwd.IAT.Mean: The mean inter-arrival time for forward packets.

Fwd.IAT.Std: The standard deviation of the inter-arrival times for forward packets.

Fwd.IAT.Max: The maximum inter-arrival time for forward packets.

Fwd.IAT.Min: The minimum inter-arrival time for forward packets.

Bwd.IAT.Total: The total inter-arrival time for backward packets.

Bwd.IAT.Mean: The mean inter-arrival time for backward packets.

Bwd.IAT.Std: The standard deviation of the inter-arrival times for backward packets.

Bwd.IAT.Max: The maximum inter-arrival time for backward packets.

Bwd.IAT.Min: The minimum inter-arrival time for backward packets.

Fwd.PSH.Flags: The number of packets with the PSH (Push) flag set in the forward direction.

Fwd.Header.Length: The total length of headers in forward packets.

Bwd.Header.Length: The total length of headers in backward packets.

Fwd.Packets.s: The rate of forward packets per second.

Bwd.Packets.s: The rate of backward packets per second.

Min.Packet.Length: The minimum length of any packet in the flow.

Max.Packet.Length: The maximum length of any packet in the flow.

Packet.Length.Mean: The mean (average) length of all packets in the flow.

Packet.Length.Std: The standard deviation of the lengths of all packets in the flow.

Packet.Length.Variance: The variance of the lengths of all packets in the flow.

FIN.Flag.Count: The count of packets with the FIN flag set.

SYN.Flag.Count: The count of packets with the SYN flag set.

RST.Flag.Count: The count of packets with the RST flag set.

PSH.Flag.Count: The count of packets with the PSH flag set.

ACK.Flag.Count: The count of packets with the ACK flag set.

URG.Flag.Count: The count of packets with the URG flag set.

ECE.Flag.Count: The count of packets with the ECE flag set.

Down.Up.Ratio: The ratio of download (forward) to upload (backward) traffic.

Average.Packet.Size: The average size of packets in the flow.

Avg.Fwd.Segment.Size: The average size of forward segments in the flow.

Avg.Bwd.Segment.Size: The average size of backward segments in the flow.

Fwd.Header.Length.1: The total length of forward headers in another format.

Subflow.Fwd.Packets: The number of subflow forward packets.

Subflow.Fwd.Bytes: The number of subflow forward bytes.

Subflow.Bwd.Packets: The number of subflow backward packets.

Subflow.Bwd.Bytes: The number of subflow backward bytes.

Init_Win_bytes_forward: The initial window size in bytes for the forward direction.

Init_Win_bytes_backward: The initial window size in bytes for the backward direction.

act_data_pkt_fwd: The number of actual data packets in the forward direction.

min_seg_size_forward: The minimum segment size in the forward direction.

Active.Mean: The mean time that the flow was active.

Active.Std: The standard deviation of the time that the flow was active.

Active.Max: The maximum time that the flow was active.

Active.Min: The minimum time that the flow was active.

Idle.Mean: The mean idle time between flows.

Idle.Std: The standard deviation of idle time between flows.

Idle.Max: The maximum idle time between flows.

Idle.Min: The minimum idle time between flows.

L7Protocol: The Layer 7 (application layer) protocol classification.

ProtocolName: The name of the network protocol.

```python
# Function to remove outliers using IQR
def remove_outliers(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# List of columns for which you want to remove outliers
columns_to_check = [  'Source.Port', 'Destination.Port',
       'Protocol', 'Flow.Duration', 'Total.Fwd.Packets',
       'Total.Backward.Packets', 'Total.Length.of.Fwd.Packets',
       'Total.Length.of.Bwd.Packets', 'Fwd.Packet.Length.Mean',
       'Bwd.Packet.Length.Min', 'Flow.Bytes.s', 'Flow.Packets.s',
       'Flow.IAT.Min', 'Fwd.IAT.Min', 'Bwd.IAT.Min', 'Bwd.Packets.s',
       'Min.Packet.Length', 'Packet.Length.Mean', 'Down.Up.Ratio',
       'Init_Win_bytes_forward', 'Init_Win_bytes_backward',
       'min_seg_size_forward', 'Active.Mean']  # Replace with your column names

# Remove outliers from the DataFrame
df_cleaned = remove_outliers(df, columns_to_check)

# Now, df_cleaned contains your DataFrame with outliers removed.
df_cleaned.head()
print(df_cleaned.shape)
```
